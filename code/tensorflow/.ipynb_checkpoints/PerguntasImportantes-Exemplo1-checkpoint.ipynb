```python
import pandas as pd
import tensorflow as tf
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## Rede da figura 1c

O que espero ver é que as ligações e bias relativos ao component y sejam proximos de 0.

```python
def gen_data(n):
    X = []
    y = []

    ## Gera n ponto equidistantes no intervalo [-2, 2]

    for i in np.linspace(-2, 2, n):
        ## A
        X.append( (-1, i) )
        y.append([1, 0])

        ## B
        X.append( (1, i) )
        y.append([0, 1]) 

    return np.array(X), np.array(y)
```

```python
X, y = gen_data(30)

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
X, y = gen_data(30)

weights_l0_1 = []
bias_l0_1 = []
acc_1 = []

for _ in range(500):

    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(
            2,
            activation="sigmoid"
        )
    ])

    model.compile(
        loss='mean_squared_error',
        metrics=["accuracy"]
    )

    hist = model.fit(X, y, epochs=60, batch_size=3, verbose=0)
    layer0 = model.layers[0]

    weights_l0_1.append(layer0.weights[0].numpy())
    bias_l0_1.append(layer0.weights[1].numpy())
    acc_1.append(hist.history["accuracy"][-1])
    
weights_l0_1 = np.array(weights_l0_1)
bias_l0_1 = np.array(bias_l0_1)
acc_1 = np.array(acc_1)
```

```python
plt.figure(figsize=(16, 9))


plt.subplot(221)
plt.title("w_00")
plt.hist(weights_l0_1[:, 0, 0])

plt.subplot(222)
plt.title("w_01")
plt.hist(weights_l0_1[:, 0, 1])

plt.subplot(223)
plt.title("w_10")
plt.hist(weights_l0_1[:, 1, 0])

plt.subplot(224)
plt.title("w_11")
plt.hist(weights_l0_1[:, 1, 1])

plt.show()
```

## Nuvem de pontos

Ao usar uma nuvem de pontos é esperado que o comportamento seja igual, já que novamente a compoente que importa é a x.

```python
def gen_data_cloud(n):
    X = []
    y = []
    
    y_axis = np.linspace(-2, 2, n)
    x1 = np.linspace(-1.5, -0.5, n)
    x2 = np.linspace(0.5, 1.5, n)

    x1v, y1v = np.meshgrid(x1, y_axis)
    x2v, y2v = np.meshgrid(x2, y_axis)

    for i in range(n):
        for j in range(n):
            
            _x1 = x1v[j,i]
            _y1 = y1v[j,i]

            _x2 = x2v[j,i]
            _y2 = y2v[j,i]

            X.append( (_x1, _y1) )
            y.append([1, 0])
            
            X.append( (_x2, _y2) )
            y.append([0, 1])

    return np.array(X), np.array(y)
```

Visualizando os dados

```python
X, y = gen_data_cloud(15)

y_class = np.argmax(y, axis=1)
c = []

for i in y_class:
    if i == 0:
        c.append("b")
    else:
        c.append("g")


plt.scatter(X[:, 0], X[:, 1], c=c)
plt.show()
```

Rede neural do desenho 1c

```python
X, y = gen_data_cloud(15)


weights_cloud_l0_1 = []
bias_cloud_l0_1 = []
acc_cloud_1 = []

for i in range(500):
    print("Treinando rede {}..".format(i))
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(
            2,
            activation="sigmoid"
        )
    ])

    model.compile(
        loss='mean_squared_error',
        metrics=["accuracy"]
    )

    hist = model.fit(X, y, epochs=60, batch_size=3, verbose=0)
    layer0 = model.layers[0]

    weights_cloud_l0_1.append(layer0.weights[0].numpy())
    bias_cloud_l0_1.append(layer0.weights[1].numpy())
    acc_cloud_1.append(hist.history["accuracy"][-1])
    
    
weights_cloud_l0_1 = np.array(weights_cloud_l0_1)
bias_cloud_l0_1 = np.array(bias_cloud_l0_1)
acc_cloud_1 = np.array(acc_cloud_1)
```

```python
plt.figure(figsize=(16, 9))


plt.subplot(221)
plt.title("w_00")
plt.hist(weights_cloud_l0_1[:, 0, 0])

plt.subplot(222)
plt.title("w_01")
plt.hist(weights_cloud_l0_1[:, 0, 1])

plt.subplot(223)
plt.title("w_10")
plt.hist(weights_cloud_l0_1[:, 1, 0])

plt.subplot(224)
plt.title("w_11")
plt.hist(weights_cloud_l0_1[:, 1, 1])

plt.show()
```

novamente os pesos relacionados com a entrada 1 são mais "fortes". Adicionado um neurônio é esperado que o comportamênto permaneça igual.

```python
def gen_data_cloud_2(n):
    X = []
    y = []
    
    y_axis = np.linspace(-2, 2, n)
    x1 = np.linspace(-1.5, -0.5, n)
    x2 = np.linspace(0.5, 1.5, n)

    x1v, y1v = np.meshgrid(x1, y_axis)
    x2v, y2v = np.meshgrid(x2, y_axis)

    for i in range(n):
        for j in range(n):
            
            _x1 = x1v[j,i]
            _y1 = y1v[j,i]

            _x2 = x2v[j,i]
            _y2 = y2v[j,i]

            X.append( (_x1, _y1, _y1) )
            y.append([1, 0])
            
            X.append( (_x2, _y2, _y2) )
            y.append([0, 1])

    return np.array(X), np.array(y)
```

```python
X, y = gen_data_cloud_2(15)


weights_cloud_l0_2 = []
bias_cloud_l0_2 = []
acc_cloud_2 = []

for i in range(500):
    print("Training model {}...".format(i))
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(
            2,
            activation="sigmoid"
        )
    ])

    model.compile(
        loss='mean_squared_error',
        metrics=["accuracy"]
    )

    hist = model.fit(X, y, epochs=60, batch_size=3, verbose=0)
    layer0 = model.layers[0]

    print("\t accuracy: {}".format(hist.history["accuracy"][-1]))
    weights_cloud_l0_2.append(layer0.weights[0].numpy())
    bias_cloud_l0_2.append(layer0.weights[1].numpy())
    acc_cloud_2.append(hist.history["accuracy"][-1])
    
    
weights_cloud_l0_2 = np.array(weights_cloud_l0_2)
bias_cloud_l0_2 = np.array(bias_cloud_l0_2)
acc_cloud_2 = np.array(acc_cloud_2)
```

```python
plt.figure(figsize=(16, 16))

plt.subplot(321)
plt.title("w_00")
plt.hist(weights_cloud_l0_2[:, 0, 0])

plt.subplot(322)
plt.title("w_01")
plt.hist(weights_cloud_l0_2[:, 0, 1])

plt.subplot(323)
plt.title("w_10")
plt.hist(weights_cloud_l0_2[:, 1, 0])

plt.subplot(324)
plt.title("w_11")
plt.hist(weights_cloud_l0_2[:, 1, 1])

plt.subplot(325)
plt.title("w_20")
plt.hist(weights_cloud_l0_2[:, 2, 0])

plt.subplot(326)
plt.title("w_21")
plt.hist(weights_cloud_l0_2[:, 2, 1])

plt.show()
```
