```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import gudhi as gd
import seaborn as sns

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

sns.set()
```

```python
def data_2(n = 100):
    x = np.linspace(-1, 1, n)
    y_1 = x ** 2 - 0.2
    y_2 = x ** 2 - 1
    
    x1 = list(zip(x, y_1))
    x2 = list(zip(x, y_2))
    
    y1 = [ [0, 1] for _ in range(n) ]
    y2 = [ [1, 0] for _ in range(n) ]
    
    X = x1 + x2
    y = y1 + y2
    
    return np.array(X), np.array(y)
```

```python
X, y = data_2(500)

c_max = np.argmax(y, axis=1)
c = []

for i in c_max:
    if i == 0:
        c.append("g")
    else:
        c.append("r")

plt.scatter(X[:, 0], X[:, 1], c=c)
plt.show()
```

# Model

```python
model = Sequential([
    Dense(2, activation="sigmoid"),
    Dense(2, activation="sigmoid")
])

model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.05),
    metrics=["accuracy"]
)

hist = model.fit(X, y, epochs=10, batch_size=12)
```

Uma pergunta interessante porque rede com uma cada 2 neuronios de entrada e 2 de saida nao conseguem classificar os pontos acima corretamente?

```python
w1 = model.layers[0].weights[0].numpy()
b1 = model.layers[0].weights[1].numpy()

w2 = model.layers[1].weights[0].numpy()
b2 = model.layers[1].weights[1].numpy()


a1 = tf.keras.activations.sigmoid( X.dot(w1) + b1 )
a2 = tf.keras.activations.sigmoid( a1.numpy().dot(w2) + b2 )


a1 = a1.numpy()
a2 = a2.numpy()

plt.title("Output camada intermediaria")
plt.scatter(a1[:, 0], a1[:, 1])
plt.show()

plt.title("Output camada de saida")
plt.scatter(a2[:, 0], a2[:, 1])
plt.show()
```

```python
skeleton = gd.RipsComplex(points = a2, max_edge_length = 1.1)

Rips_simplex_tree = skeleton.create_simplex_tree(max_dimension = 0)

BarCodes_Rips0 = Rips_simplex_tree.persistence()

gd.plot_persistence_barcode(BarCodes_Rips0);
```

# Rede ruim

```python
model = Sequential([
    Dense(2, activation="sigmoid")
])

model.compile(
    loss="mean_squared_error",
    metrics=["accuracy"]
)

hist = model.fit(X, y, epochs=10, batch_size=12)
```

```python
w1 = model.layers[0].weights[0].numpy()
b1 = model.layers[0].weights[1].numpy()

a1 = tf.keras.activations.sigmoid( X.dot(w1) + b1 )
a1 = a1.numpy()

plt.title("Output camada intermediaria")
plt.scatter(a1[:, 0], a1[:, 1])
plt.show()
```

```python
skeleton = gd.RipsComplex(points = a1, max_edge_length = 0.8)

Rips_simplex_tree = skeleton.create_simplex_tree(max_dimension = 0)

BarCodes_Rips0 = Rips_simplex_tree.persistence()

gd.plot_persistence_barcode(BarCodes_Rips0);
```
